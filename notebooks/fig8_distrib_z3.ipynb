# make (and save) plots: 

# z=3 | M$_*$, SFR, A$_V$ distribution [violin plots]

```python
fig, axes = joypy.joyplot(data = z3_cat_mass,linecolor='w',ylabels=True, xlabelsize=20, ylabelsize=20,
                        figsize=(7,12),grid=True,yrot=0,colormap=matplotlib.cm.viridis_r)
for i in range(len(axes)):
    tempy = axes[i].set_ylim()
    axes[i].set_ylim(tempy[0],tempy[1]+1.0)
    axes[i].set_xlim(8,12)
plt.xlabel(r'log(M/M$_\odot$)')
plt.savefig("../figures/fig8_distrib_z3_mass.png", dpi=600, bbox_inches='tight')

plt.show()
```
