## main sequence

```python
#plot main sequence with median mass and SFR
z3_cat_sfr_50 = np.nanmedian(z3_cat_sfr,axis=1)
z3_cat_mass_50 = np.nanmedian(z3_cat_mass,axis=1)
z3_cat_Av_50 = np.nanmedian(z3_cat_av,axis=1)

plt.figure(figsize=[8,5])
plt.scatter(z3_cat_mass_50,z3_cat_sfr_50,c=z3_cat_Av_50,cmap='rainbow')
plt.xlim(8.5,12)
plt.ylim(-2.5,3.5)
plt.clim(0,2)
plt.colorbar(label='median A$_v$')
plt.title("z~3")
plt.xlabel(r'median log(M/M$_{\odot}$)')
plt.ylabel(r'median log[SFR/(M$_\odot$yr$^{-1}$)]')
plt.savefig("../figures/fig10_ms_z3.png", dpi=600, bbox_inches='tight')
plt.show()

coeff = np.polyfit(z3_cat_mass_50[z3_cat_sfr_50-z3_cat_mass_50>-10],z3_cat_sfr_50[z3_cat_sfr_50-z3_cat_mass_50>-10],1)
#print(coeff)

xx = np.arange(8,13,0.1)
#yy = coeff[3] + coeff[2]*xx + coeff[1]*xx**2 + coeff[0]*xx**3
yy = coeff[1] + coeff[0]*xx 

#average scatter?
liney = coeff[1] + coeff[0]*z3_cat_mass_50[z3_cat_sfr_50-z3_cat_mass_50>-10] 
diffyy = abs(z3_cat_sfr_50[z3_cat_sfr_50-z3_cat_mass_50>-10] - liney)
meandiff = np.mean(diffyy)
print(meandiff)


plt.figure(figsize=[8,5])
plt.scatter(z3_cat_mass_50,z3_cat_sfr_50,color='gray')
plt.scatter(z3_cat_mass_50[z3_cat_sfr_50-z3_cat_mass_50>-10],z3_cat_sfr_50[z3_cat_sfr_50-z3_cat_mass_50>-10],color='blue')
plt.plot(xx,yy,color='darkorange',linewidth=3,linestyle='-')
plt.plot(xx,yy+meandiff,color='darkorange',linewidth=3,linestyle='--')
plt.plot(xx,yy-meandiff,color='darkorange',linewidth=3,linestyle='--')
plt.xlim(8.5,12)
plt.ylim(-2.5,3.5)
#plt.title("y = -6.15 +0.76x")
plt.xlabel(r'median log(M/M$_{\odot}$)')
plt.ylabel(r'median log[SFR/(M$_\odot$yr$^{-1}$)]')
plt.savefig("../figures/fig10_ms_z3_fit.png", dpi=600, bbox_inches='tight')
plt.show()
```
