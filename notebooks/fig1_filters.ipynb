# Figure with SED and filters

```python
#plotting and tabling
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)

params={'legend.fontsize':'18','axes.labelsize':'18',
        'axes.titlesize':'18','xtick.labelsize':'18',
        'ytick.labelsize':'18','lines.linewidth':2,
        'axes.linewidth':2,'animation.html': 'html5',
        'figure.figsize':(8,6)}
plt.rcParams.update(params)
plt.rcParams.update({'figure.max_open_warning': 0})
import matplotlib.patches as patches

import numpy as np
import pandas as pd
```

## Read one of the template spectra for LEGA-C

```python
filesed = '../templates/00005.dat'
sed = pd.read_table(filesed,sep='\s+',header=None)
sed.head()
```

## Read each filter

```python
filt1 = '../filters/CTIO_U.dat'
filt2 = '../filters/ACS_F435W.dat'
filt3 = '../filters/ACS_F606W.dat'
filt4 = '../filters/ACS_F775W.dat'
filt5 = '../filters/ACS_F814W.dat'
filt6 = '../filters/ACS_F850LP.dat'
filt7 = '../filters/WFC3_F098M.dat'
filt8 = '../filters/WFC3_F105W.dat'
filt9 = '../filters/WFC3_F125W.dat'
filt10 = '../filters/WFC3_F160W.dat'
filt11 = '../filters/HAWKI_KS.dat'
filt12 = '../filters/IRAC_CH1.dat'
filt13 = '../filters/IRAC_CH2.dat'

f1 = pd.read_table(filt1,sep='\s+',comment='#',header=None)
f2 = pd.read_table(filt2,sep='\s+',comment='#',header=None)
f3 = pd.read_table(filt3,sep='\s+',comment='#',header=None)
f4 = pd.read_table(filt4,sep='\s+',comment='#',header=None)
f5 = pd.read_table(filt5,sep='\s+',comment='#',header=None)
f6 = pd.read_table(filt6,sep='\s+',comment='#',header=None)
f7 = pd.read_table(filt7,sep='\s+',comment='#',header=None)
f8 = pd.read_table(filt8,sep='\s+',comment='#',header=None)
f9 = pd.read_table(filt9,sep='\s+',comment='#',header=None)
f10 = pd.read_table(filt10,sep='\s+',comment='#',header=None)
f11 = pd.read_table(filt11,sep='\s+',comment='#',header=None)
f12 = pd.read_table(filt12,sep='\s+',comment='#',header=None)
f13 = pd.read_table(filt13,sep='\s+',comment='#',header=None)
```

# Plot SED and filters

```python
zz1 = (1. + 1.)
zz3 = (1. + 3.)

plt.plot(sed[0]/1E4 * zz1,np.log10(sed[1]/zz1)-3.5, color='darkgray')
plt.plot(sed[0]/1E4 * zz3,np.log10(sed[1]/zz3)-4.5, color='darkgray')

plt.plot(f1[0]/1E4,f1[1],label='U',color='magenta')
plt.plot(f2[0]/1E4,f2[1],label='F435W',color='blue')
plt.plot(f3[0]/1E4,f3[1],label='F606W',color='dodgerblue')
plt.plot(f4[0]/1E4,f4[1],label='F775W',color='seagreen')
plt.plot(f5[0]/1E4,f5[1],label='F814W',color='darkgreen')
plt.plot(f6[0]/1E4,f6[1],label='F850LP',color='darkolivegreen')
plt.plot(f7[0]/1E4,f7[1],label='F098M',color='gold')
plt.plot(f8[0]/1E4,f8[1],label='F105W',color='darkorange')
plt.plot(f9[0]/1E4,f9[1],label='F125W',color='peru')
plt.plot(f10[0]/1E4,f10[1],label='F160W',color='orangered')
plt.plot(f11[0]/1E4,f11[1]/100,label='K',color='brown')
plt.plot(f12[0]/1E4,f12[1],label='IRAC Ch1',color='darkred')
plt.plot(f13[0]/1E4,f13[1],label='IRAC Ch2',color='black')

plt.xscale('log')

xticks = [0.3, 1, 3]
xlabels = [0.3, 1, 3]

plt.xticks(xticks,xlabels)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlim(0.25, 7)
plt.ylim(0,4)

plt.xlabel("wavelength ($\mu$m)")
plt.ylabel("flux normalized")

plt.savefig("../figures/fig1_filters.png", dpi=600, bbox_inches='tight')

plt.show()
```
